```python
### data loading to database
def connection(dbname, host, port, user, password):
    conn = psycopg2.connect(dbname=dbname, host=host, port=port, user=user, password=password)
    return conn

conn = connection('kikee','localhost',5432,'postgres','Izuchi2004$@') 

def loading(file_path, table_name, conn):
    with conn.cursor() as cur:
        with open(file_path, 'r') as f:
            next(f)
            cur.copy_expert(
                f"copy {table_name} from STDIN WITH CSV",
                f
            )
        conn.commit()
    print(f"Loaded file into {table_name} using copy.")


file_path = ['data\order_table.csv','data\product_table.csv','data\sales_table.csv']
table_name = ['order_table','product_table','sale_table']
for path, table in zip(file_path,table_name):
    loading(path, table, conn)
```
